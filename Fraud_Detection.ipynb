## 1. ðŸ¤– Machine Learning Overview

## 2. ðŸŒ³ Decision Tree Algorithm

A **Decision Tree** is a supervised learning algorithm that makes predictions  
by learning **simple decision rules** from the data features.  
It splits the data recursively based on the most informative feature at each node.
```
          [Amount > 5000 ?]
           /              \
         OUI              NON
          â”‚                â”‚
  [Type == TRANSFER ?]   No Fraud âœ…
    /           \
  OUI           NON
   â”‚             â”‚
Fraud ðŸš¨      No Fraud âœ…
```

---

## 3. âš™ï¸ How the Tree is Built

The algorithm selects at each node the feature that best separates the classes,  
using one of these criteria :

### Gini Impurity *(default)*
$$Gini = 1 - \sum_{i=1}^{n} p_i^2$$

> Measures the **impurity** of a node.  
> A node is **pure** when Gini = 0 (only one class).

### Entropy
$$Entropy = - \sum_{i=1}^{n} p_i \cdot \log_2(p_i)$$

> Measures the **disorder** in a node.  
> Based on **Information Theory**.

---

## 4. ðŸ”‘ Key Concepts

| Concept | Definition |
|---|---|
| **Root Node** | First split â€” most important feature |
| **Internal Node** | Intermediate decision point |
| **Leaf Node** | Final decision â†’ Fraud or No Fraud |
| **Depth** | Number of levels in the tree |
| **Pruning** | Reducing tree size to avoid overfitting |

---

## 5. ðŸ“Š Evaluation Metrics

Since the dataset is **highly imbalanced** (0.13% Fraud),  
Accuracy alone is not sufficient. We use :

### Precision
$$Precision = \frac{TP}{TP + FP}$$
> Among all predicted frauds, how many were **really fraud** ?

### Recall
$$Recall = \frac{TP}{TP + FN}$$
> Among all real frauds, how many did the model **detect** ?

### F1-Score
$$F1 = 2 \times \frac{Precision \times Recall}{Precision + Recall}$$
> Harmonic mean between Precision and Recall.  
> **Best metric for imbalanced datasets.**

### Confusion Matrix
```
                  Predicted No Fraud    Predicted Fraud
Actual No Fraud  |   True Negative   |  False Positive  |
Actual Fraud     |   False Negative  |  True Positive   |
